```python
data[data[" Flow Packets/s"]==float("inf")][" Label"].value_counts()
```

```python
for n,m in data[[" Flow Packets/s", " Label"]].iterrows():
  if n == float('inf') and m == 'Benign':
    print(n,m)
```

```python
data = data.drop(df[df[" Flow Packets/s"] == float('inf')].index)
```

```python
data[data[" Flow Packets/s"]==float("inf")][" Label"].value_counts()
```
